```python
import numpy as np
import termcolor
```

```python
def load_data(fname = "data.sample.txt"):
    
    with open(fname, "r") as f:
        
        lines = f.readlines()
    
    data = []
    
    for line in lines:
        
        index, sent1, sent2, _, _, label = line.strip().split("\t")
        index, label = int(index.split(".")[0]), int(label)
        sent1, sent2 = sent1.split(" "), sent2.split(" ")
        if sent1 == sent2: continue
        data.append((sent1, sent2, index, label))
        
    return data

def display_example(example):
    
    sent1, sent2, index, label = example
    color = "blue" if label == 1 else "red"
    
    sent1_before = sent1[:index]
    sent2_before = sent2[:index]
    sent1_after = sent1[index + 1:]
    sent2_after = sent2[index + 1:]
    
    w1 = "***" + termcolor.colored(sent1[index], color) + "***"
    w2 =  "***" + termcolor.colored(sent2[index], color) +  "***"
    
    sent1_str = " ".join(sent1_before) + " " + w1 + " " + " ".join(sent1_after)
    sent2_str = " ".join(sent2_before) + " " + w2 + " " + " ".join(sent2_after)
    
    s = "Label: {}\n".format(label) + sent1_str + "\n" + sent2_str + "\n==================================================\n"
    print(s)

    
```

```python
data = load_data()
for example in data:
    display_example(example)
```
